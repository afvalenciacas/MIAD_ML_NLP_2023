**limitaciones**

La regresión lineal tiene algunas limitaciones que es importante tener en cuenta. Por ejemplo, este modelo asume una relación lineal entre las variables independientes y la variable dependiente, lo cual puede no ser válido en todos los casos. Además, la regresión lineal es sensible a valores atípicos y errores de medición, lo que puede afectar la precisión de las predicciones. Otra limitación a considerar es que el modelo asume ciertos supuestos, como el de colinealidad y normalidad, los cuales pueden no cumplirse en algunos casos y afectar la validez de los resultados. Es importante tener en cuenta estas limitaciones y evaluar si la regresión lineal es adecuada para el problema en cuestión.

### Punto 4 - Árbol de decisión manual
En la celda 4 cree un árbol de decisiones para pronosticar la variable "total" iterando **manualmente** sobre las variables "hour" y  "season". El árbol debe tener al menos 6 nodos finales.

```python
# Separación de variables predictoras (X) y variable de interes (y)
X = bikes[['hour','season']]
y = bikes['total']

# Definición de la función que calcula el gini index
def gini(y):
    if y.shape[0] == 0:
        return 0
    else:
        return 1 - (y.mean()**2 + (1 - y.mean())**2)
# Definición de la función gini_imputiry para calular la ganancia de una variable predictora j dado el punto de corte k
def gini_impurity(X_col, y, split):
    
    filter_l = X_col < split
    y_l = y.loc[filter_l]
    y_r = y.loc[~filter_l]
    
    n_l = y_l.shape[0]
    n_r = y_r.shape[0]
    
    gini_y = gini(y)
    gini_l = gini(y_l)
    gini_r = gini(y_r)
    
    gini_impurity_ = gini_y - (n_l / (n_l + n_r) * gini_l + n_r / (n_l + n_r) * gini_r)
    
    return gini_impurity_

# Definición de la función best_split para calcular cuál es la mejor variable y punto de cortepara hacer la bifurcación del árbol
def best_split(X, y, num_pct=10):
    
    features = range(X.shape[1])
    
    best_split = [0, 0, 0]  # j, split, gain
    
    # Para todas las varibles 
    for j in features:
        
        splits = np.percentile(X.iloc[:, j], np.arange(0, 100, 100.0 / (num_pct+1)).tolist())
        splits = np.unique(splits)[1:]
        
        # Para cada partición
        for split in splits:
            gain = gini_impurity(X.iloc[:, j], y, split)
                        
            if gain > best_split[2]:
                best_split = [j, split, gain]
    
    return best_split

# Función ára contar el número de nodos finales en el arbol actual. Definido en almenos 6 nodos finales
def count_leaf_nodes(tree):
    if 'sl' not in tree and 'sr' not in tree:
        return 1
    else:
        return count_leaf_nodes(tree['sl']) + count_leaf_nodes(tree['sr'])

# Definición de la función tree_grow para hacer un crecimiento recursivo del árbol
def tree_grow(X, y, level=0, min_gain=0.001, max_depth=None, num_pct=10):
    
    # Si solo es una observación
    if X.shape[0] == 1:
        tree = dict(y_pred=y.iloc[:1].values[0], y_prob=0.5, level=level, split=-1, n_samples=1, gain=0)
        return tree
    
    # Calcular la mejor división
    j, split, gain = best_split(X, y, num_pct)
    
    # Guardar el árbol y estimar la predicción
    y_pred = int(y.mean() >= 0.5) 
    y_prob = (y.sum() + 1.0) / (y.shape[0] + 2.0)  # Corrección Laplace 
    
    tree = dict(y_pred=y_pred, y_prob=y_prob, level=level, split=-1, n_samples=X.shape[0], gain=gain)
    
    # Revisar el criterio de parada 
    if gain < min_gain:
        return tree
    if max_depth is not None:
        if level >= max_depth:
            return tree
    if count_leaf_nodes(tree) >= 6:
        return tree
    
    # Continuar creando la partición
    filter_l = X.iloc[:, j] < split
    X_l, y_l = X.loc[filter_l], y.loc[filter_l]
    X_r, y_r = X.loc[~filter_l], y.loc[~filter_l]
    tree['split'] = [j, split]

    # Siguiente iteración para cada partición
    
    tree['sl'] = tree_grow(X_l, y_l, level + 1, min_gain=min_gain, max_depth=max_depth, num_pct=num_pct)
    tree['sr'] = tree_grow(X_r, y_r, level + 1, min_gain=min_gain, max_depth=max_depth, num_pct=num_pct)
    
    return tree

tree_grow(X, y, level=0, min_gain=0.001, max_depth=None, num_pct=10)
```

### Punto 5 - Árbol de decisión con librería
En la celda 5 entrene un árbol de decisiones con la **librería sklearn**, usando las variables predictoras "season" y "hour" y calibre los parámetros que considere conveniente para obtener un mejor desempeño. Recuerde dividir los datos en conjuntos de entrenamiento y validación para esto. Comente el desempeño del modelo con alguna métrica de desempeño de modelos de regresión y compare desempeño con el modelo del punto 3.

```python
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

X = bikes[['hour','season']]
y = bikes['total']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

regressor = DecisionTreeRegressor(random_state=1)

# entrenamos el árbol de decisiones con los datos de entrenamiento
regressor.fit(X_train, y_train)

# hacemos las predicciones con los datos de prueba
y_train_pred = regressor.predict(X_train)
y_test_pred = regressor.predict(X_test)

mse_train = mean_squared_error(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)

print("El valor del MSE en los datos de entrenamiento es:", mse_train)
print("El valor del MSE en los datos de prueba es:", mse_test)
```

```python
#Búsqueda de los hiperparametros para el árbol
from sklearn.model_selection import GridSearchCV

reg_tree = DecisionTreeRegressor()

# Definir los valores posibles para los hiperparámetros que queremos probar
param_grid = {
    'max_depth': [None, 3, 5, 7, 9],
    'min_samples_split': [1, 2, 3, 4],
    'min_samples_leaf': [1, 2, 3, 4]
}

# Ejecutar la búsqueda en cuadrícula para encontrar los mejores hiperparámetros
grid_search = GridSearchCV(reg_tree, param_grid, cv=10)
grid_search.fit(X_train, y_train)

# Obtener los mejores hiperparámetros
best_params = grid_search.best_params_
print(f"Los mejores hiperparámetros son: {best_params}")

# Entrenar el árbol de decisión con los mejores hiperparámetros
reg_tree = DecisionTreeRegressor(**best_params)
reg_tree.fit(X_train, y_train)

# Evaluar el modelo en el conjunto de prueba
y_pred = reg_tree.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f"El error cuadrático medio en el conjunto de prueba es: {mse}")
```

### Punto 7 - Votación Mayoritaria
En la celda 7 elabore un esamble con la metodología de **Votación mayoritaria** compuesto por 300 muestras bagged para cada uno de los siguientes escenarios:

-100 árboles de decisión donde max_depth tome un valor de su elección\
-100 árboles de decisión donde min_samples_leaf tome un valor de su elección\
-100 regresiones logísticas

Evalúe cada uno de los tres modelos de manera independiente utilizando las métricas de Accuracy y F1-Score, luego evalúe el ensamble de modelos y compare los resultados. 

```python
#Definición del modelo BaggingClassifier de la libreria sklearn con Votación mayoritaria
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

#Creamos los modelos base para la votación mayoritaria
tree1 = DecisionTreeClassifier(max_depth=5)
tree2 = DecisionTreeClassifier(min_samples_leaf=10)
logreg = LogisticRegression(random_state=0)

#Creamos la instancia del Bagging Classifier para cada modelo
bagging_tree1 = BaggingClassifier(base_estimator=tree1, n_estimators=100, random_state=0)
bagging_tree2 = BaggingClassifier(base_estimator=tree2, n_estimators=100, random_state=0)
bagging_logreg = BaggingClassifier(base_estimator=logreg, n_estimators=100, random_state=0)

#Entrenamos cada modelo y evaluamos el desempeño en el conjunto de test
bagging_tree1.fit(X_train, y_train)
y_pred_tree1 = bagging_tree1.predict(X_test)
f1_score_tree1 = metrics.f1_score(y_test, y_pred_tree1)
accuracy_tree1 = metrics.accuracy_score(y_test, y_pred_tree1)

bagging_tree2.fit(X_train, y_train)
y_pred_tree2 = bagging_tree2.predict(X_test)
f1_score_tree2 = metrics.f1_score(y_test, y_pred_tree2)
accuracy_tree2 = metrics.accuracy_score(y_test, y_pred_tree2)

bagging_logreg.fit(X_train, y_train)
y_pred_logreg = bagging_logreg.predict(X_test)
f1_score_logreg = metrics.f1_score(y_test, y_pred_logreg)
accuracy_logreg = metrics.accuracy_score(y_test, y_pred_logreg)

#Evaluamos el desempeño del modelo de votación mayoritaria
y_pred_voting = (y_pred_tree1 + y_pred_tree2 + y_pred_logreg) // 2
f1_score_voting = metrics.f1_score(y_test, y_pred_voting)
accuracy_voting = metrics.accuracy_score(y_test, y_pred_voting)

#Creamos una tabla con los resultados de cada modelo y de la votación mayoritaria
results = pd.DataFrame({'Modelo': ['Árbol 1', 'Árbol 2', 'Logística', 'Votación mayoritaria'],
'F1-Score': [f1_score_tree1, f1_score_tree2, f1_score_logreg, f1_score_voting],
'Accuracy': [accuracy_tree1, accuracy_tree2, accuracy_logreg, accuracy_voting]})
print(results)
```

El modelo "Árbol 2" tuvo el mejor desempeño en ambas métricas, con un F1-Score de 0.6767 y una Accuracy de 0.6707. El segundo mejor modelo fue la votación, con un F1-Score de 0.6505 y una Accuracy de 0.6547.

El árbol 1 tuvo un desempeño inferior en comparación con los otros modelos, con un F1-Score de 0.6403 y una Accuracy de 0.6487. La regresión logística tuvo el peor desempeño en ambas métricas, con un F1-Score de 0.6053 y una Accuracy de 0.6140.

En resumen, el árbol 2 tuvo el mejor desempeño individual, pero la votación logró mejorar el desempeño general al combinar los resultados de los tres modelos.

```python
#Creamos los modelos base para la votación mayoritaria
tree1 = DecisionTreeClassifier(max_depth=5)
tree2 = DecisionTreeClassifier(min_samples_leaf=10)
logreg = LogisticRegression(random_state=0)

# Definición de las instancias del Bagging Classifier para cada modelo
bagging_tree1 = BaggingClassifier(base_estimator=tree1, n_estimators=100, random_state=0, oob_score=True)
bagging_tree2 = BaggingClassifier(base_estimator=tree2, n_estimators=100, random_state=0, oob_score=True)
bagging_logreg = BaggingClassifier(base_estimator=logreg, n_estimators=100, random_state=0, oob_score=True)

# Entrenamiento de cada modelo y evaluación de su desempeño en el conjunto de test
bagging_tree1.fit(X_train, y_train)
y_pred_tree1 = bagging_tree1.predict(X_test)
f1_score_tree1 = metrics.f1_score(y_test, y_pred_tree1)
accuracy_tree1 = metrics.accuracy_score(y_test, y_pred_tree1)

bagging_tree2.fit(X_train, y_train)
y_pred_tree2 = bagging_tree2.predict(X_test)
f1_score_tree2 = metrics.f1_score(y_test, y_pred_tree2)
accuracy_tree2 = metrics.accuracy_score(y_test, y_pred_tree2)

bagging_logreg.fit(X_train, y_train)
y_pred_logreg = bagging_logreg.predict(X_test)
f1_score_logreg = metrics.f1_score(y_test, y_pred_logreg)
accuracy_logreg = metrics.accuracy_score(y_test, y_pred_logreg)

# Obtención de los pesos alpha de cada modelo de acuerdo al error OOB
errors = np.zeros(3)
y_pred_all_ = np.zeros((X_test.shape[0], 3))

for i, bagging_model in enumerate([bagging_tree1, bagging_tree2, bagging_logreg]):
    oob_sample = ~bagging_model.estimators_samples_[i]
    y_pred_ = bagging_model.estimators_[i].predict(X_train.values[oob_sample])
    errors[i] = metrics.accuracy_score(y_pred_, y_train.values[oob_sample])
    y_pred_all_[:, i] = bagging_model.estimators_[i].predict(X_test)

alpha = (1 - errors) / (1 - errors).sum()

# Evaluación del desempeño del modelo de votación ponderada
y_pred_voting = np.sum(y_pred_all_ * alpha, axis=1) >= 0.5
f1_score_voting = metrics.f1_score(y_test, y_pred_voting)
accuracy_voting = metrics.accuracy_score(y_test, y_pred_voting)

# Creación de una tabla con los resultados de cada modelo y de la votación ponderada
results = pd.DataFrame({'Modelo': ['Árbol 1', 'Árbol 2', 'Logística', 'Votación ponderada'],
                        'F1-Score': [f1_score_tree1, f1_score_tree2, f1_score_logreg, f1_score_voting],
                        'Accuracy': [accuracy_tree1, accuracy_tree2, accuracy_logreg, accuracy_voting]})
print(results)
```

Sin embargo, la votación mayoritaria tiene un F1-Score ligeramente superior y una precisión similar a la votación ponderada. Esto indica que la votación mayoritaria es una buena opción para combinar las predicciones de diferentes modelos y mejorar el rendimiento general del modelo de ensamblaje.

Por otro lado, la votación ponderada otorga pesos diferentes a las predicciones de los modelos en función de su rendimiento individual, lo que puede ser útil en situaciones en las que algunos modelos tienen un rendimiento notablemente mejor que otros. Sin embargo, en este caso en particular, la votación ponderada no parece mejorar significativamente el rendimiento en comparación con la votación mayoritaria.

**Ventajas y desventajas**

La principal ventaja de la votación mayoritaria es que es fácil de entender e implementar, ya que simplemente se selecciona la clase que obtiene la mayoría de votos. Sin embargo, una desventaja de la votación mayoritaria es que todos los modelos tienen el mismo peso en la votación, lo que significa que un modelo incorrecto puede tener un impacto negativo en la precisión de la votación general. En este caso el peor modelo no tienen unas medidas demasiado bajas que incurran en algun error y sesguen el modelo ensamblado.

En cuanto a la votación ponderada una ventaja es que cada modelo puede contribuir con una cantidad diferente de peso a la votación final, lo que significa que los modelos más precisos tendrán un impacto mayor en el resultado final. Esto puede mejorar la precisión de la votación general. Sin embargo, una desventaja de la votación ponderada es que su implementación puede ser más complicada que la de la votación mayoritaria, ya que se requiere la selección de los pesos adecuados para cada modelo. Además, la votación ponderada puede ser sensible a valores atípicos o errores en los pesos asignados a los modelos.
